```
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

The data set is loaded and  save it x and y variables
ploting the data with loaded variables

```
data = np.loadtxt("ex1data2.txt",delimiter = ',')
x = data[:,0:2]
y = data[:,2]
plt.scatter(x[:,0],y,marker = '*',color = 'r')
plt.title("area vs price ")
plt.xlabel("area")
plt.ylabel("price($)")
plt.figure(2)
plt.title("no_of_beds vs price")
plt.scatter(x[:,1],y,marker = '.',color = 'r')
plt.xlabel("no_of_beds")
plt.ylabel("prices($)")
#normalizing the input 
x[:,0] = x[:,0]/4500
x[:,1] = x[:,1]/5
y = y/700000
```

```
model = keras.Sequential([keras.layers.Dense(units = 1,input_shape = [2])])
model.compile(optimizer='sgd',loss = 'mean_squared_error')
model.fit(x,y,epochs = 500)
```

```
y_preict = model.predict(x)
```
